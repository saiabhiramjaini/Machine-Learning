#### Matplotlib is a powerful plotting library in Python that allows you to create a wide variety of static, animated, and interactive visualizations. It's particularly useful for creating plots and graphs to visualize data.

**importing matplotlib**

```python
import matplotlib.pyplot as plt
```

```python
# Months and corresponding temperatures
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
temperatures = [30, 32, 45, 50, 60, 70, 75, 73, 65, 55, 45, 35]

# Plotting the data
plt.plot(months, temperatures)

# Adding titles and labels
plt.title('Monthly Temperatures')
plt.xlabel('Months')
plt.ylabel('Temperature (Â°F)')

# Display the plot
plt.show()
```

#### Explanation of Customizations

- **Color:** `color='blue'` sets the line color to blue.

- **Marker:** `marker='o'` uses circle markers for each data point.

- **Line Style:** `linestyle='dashed'` makes the line dashed.

- **Line Width:** `linewidth=2` sets the width of the line.

- **Marker Size:** `markersize=10` sets the size of the markers.

- **Grid:** `plt.grid(True)` adds gridlines to the plot.

- **X co-ordinate limits**: `plt.xlim(0,10)`  Lower Limit, Upper Limit

- **Y co-ordinate limits**: `plt.ylim(0,20)`  Lower Limit, Upper Limit

- **Save graph**: `plt.savefig('temperature.png')`

- **legend**: `plt.legend()`tells what each line or data point represents.


```python
# Creating a bar plot
plt.bar(months, temperatures, color='skyblue')
```

bins: This parameter controls the number of bins (or intervals) into which the data is divided in the histogram. It can be an integer specifying the number of bins to create, or a sequence defining the edges of the bins. For example, bins=5 creates 5 bins. 

```python
plt.hist(temperatures, bins=20, color='green', alpha=0.7)  # Adjust bins as per your data range and distribution
```

```python
import numpy as np
```

```python
# Define the slope (m) and y-intercept (c)
m = 2
c = 1

# Generate x values
x = np.linspace(-5, 5, 100)

# Calculate y values
y = m * x + c

# Plot the line
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'y = {m}x + {c}')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of a Linear Line')
plt.grid(True)
plt.legend()

# Show plot
plt.show()
```

```python
# Define coefficients a, b, and c for the quadratic equation y = ax^2 + bx + c
a = 1
b = -2
c = 1

# Generate x values
x = np.linspace(-5, 5, 400)

# Calculate y values
y = a * x**2 + b * x + c

# Plot the quadratic curve
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'y = {a}x^2 + {b}x + {c}')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of a Quadratic Equation')
plt.grid(True)
plt.legend()

# Show plot
plt.show()
```

`plt.subplots()` in matplotlib is a function that creates a figure and a set of subplots. It's a convenient way to create multiple plots within the same figure, allowing you to organize and compare different aspects of your data or models. ~`

```python
# Example data
x = np.linspace(-5, 5, 100)
y1 = x ** 2
y2 = np.sin(x)

# Create a figure and subplots
fig, axs = plt.subplots(2, 1, figsize=(8, 10))  # Create 2 subplots vertically

# Plot on the first subplot (upper plot)
axs[0].plot(x, y1, color='blue', label='y = x^2')
axs[0].set_title('Plot of y = x^2')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].grid(True)
axs[0].legend()

# Plot on the second subplot (lower plot)
axs[1].plot(x, y2, color='red', label='y = sin(x)')
axs[1].set_title('Plot of y = sin(x)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
axs[1].grid(True)
axs[1].legend()

# Adjust layout and spacing
plt.tight_layout()

# Show the plots
plt.show()
```

to plot two graphs (or lines) in a single plane using matplotlib, you can do so by creating a single subplot with multiple lines plotted on it. 

```python
# Example data
x = np.linspace(-5, 5, 100)
y1 = x ** 2
y2 = np.sin(x)

# Plot both graphs on the same plane
plt.figure(figsize=(8, 6))

# Plot the first graph (y = x^2)
plt.plot(x, y1, color='blue', label='y = x^2')

# Plot the second graph (y = sin(x))
plt.plot(x, y2, color='red', label='y = sin(x)')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Two Graphs on a Single Plane')
plt.grid(True)
plt.legend()

# Show plot
plt.show()
```
